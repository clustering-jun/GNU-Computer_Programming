## 변수(variable) 활용하여 오륜기 그리기

```python
import turtle
t = turtle.Turtle()
t.shape("turtle")
t.pensize(10)
t.speed(0)

# 변수 r에 값 100을 저장
# 주기억 장치의 어느 주소에 100이라는 값을 저장 (관리: OS)
# 'r과 100이 같다'라는 뜻이 아님에 주의
# 이제 r(반지름)을 지정하여 오륜기를 그릴 수 있음
r = 100

t.color('black')
t.circle(r)

t.teleport(-2.2*r,0)
t.color('blue')
t.circle(r)

t.teleport(2.2*r,0)
t.color('red')
t.circle(r)

t.teleport(-1.1*r,-r)
t.color('yellow')
t.circle(r)

t.teleport(1.1*r,-r)
t.color('green')
t.circle(r)
```

## 변수(variable)에 대하여
- 변수(variable): 처리할 자료나 결과값을 임시로 보관하는 공간에 대한 논리적 표현
- 변수를 사용하여 주기억 장치의 임의의 공간을 할당 받고 자료를 저장함

image.png

```python
a = 100
print(a) # 100

print("a") # a

a = 200
print(a) # 200

b = 300
print(b) # 300

print(a, b) # 200 300

c = a + b
print(c) # 500

print(a, b, c) # 200 300 500

print("수1:",a, "수2:",b, "합:",c) # 수1: 200 수2: 300 합: 500
```

## 대입연산자(=)와 비교연산자(==)
1. **대입연산자(=)**
- 변수를 선언하거나 값을 할당할 때 사용
- 왼쪽 피연산자(변수)에 오른쪽 피연산자의 값을 저장


2. **비교연산자(==)**
- 두 개의 값이 같은지를 비교할 때 사용
- 두 값이 같으면 True, 다르면 False를 반환



```python
a = 10  # 변수 a에 10을 할당
b = "Hello"  # 변수 b에 문자열 "Hello"를 할당


x = 5
y = 10
print(x == y)  # False (5와 10은 같지 않음)

z = 10
print(y == z)  # True (10과 10은 같음)
```

```python
su = 100

print(su)  # 100

su = su + 10
print(su)  # 110
```

```python
c = c
c = c + 1 # name 'c' is not defined -> c = 0으로 두고 해야함
```

```python
c = 0
c = c + 1  # 우변의 값이 좌변에 저장됨.
print(c)
```

## 사용자가 원하는 크기의 오륜기 그리기
- 사용자가 원하는 크기의 오륜기를 그려주는 프로그램으로 업그레이드!!

```python
import turtle
t = turtle.Turtle()
t.shape("turtle")
t.pensize(10)
t.speed(0)

r = int(input('반지름: '))

t.color('black')
t.circle(r)

t.teleport(-2.2*r,0)
t.color('blue')
t.circle(r)

t.teleport(2.2*r,0)
t.color('red')
t.circle(r)

t.teleport(-1.1*r,-r)
t.color('yellow')
t.circle(r)

t.teleport(1.1*r,-r)
t.color('green')
t.circle(r)
```

## 내가 원하는 크기와 색의 원 그리기
- 사용자로부터 원의 반지름과 색을 입력받아 radius과 color라는 변수에 각각 저장하고, 해당 크기와 색의 원을 그려주는 프로그램을 작성하라.
(단, 반지름은 양의 정수만 입력해서 테스트한다.)

```python
import turtle
t = turtle.Turtle()
t.shape("turtle")
t.pensize(10)
t.speed(0)

# 사용자로부터 반지름(radius)과 색(color)을 입력받음.
radius = int(input('반지름을 입력해라: '))
color = input('색을 입력해라: ')

t.color(color)

t.begin_fill()

t.circle(radius)

t.end_fill()
```

## 데이터 입력과 데이터의 사용
- 사용자의 입력 요구에 맞는 정사각형을 그려주는 프로그램을 작성하시오.
- 한 변의 길이, 선의 두께, 선 색, 면 색 정보를 입력받는다.

```python
import turtle
t = turtle.Turtle()
t.shape("turtle")
t.speed(0)

# 입력
길이 = int(input('길이 입력: ')) #int
두께 = int(input('두께 입력: ')) #int
선색 = input('선색 입력: ') #str
면색 = input('면색 입력: ') #str

# 그리기를 위한 기본설정
t.fillcolor(면색)
t.pencolor(선색)
t.pensize(두께)

# 정사각형 그리기
t.begin_fill()

t.fd(길이)
t.rt(90)

t.fd(길이)
t.rt(90)

t.fd(길이)
t.rt(90)

t.fd(길이)
t.rt(90)

t.end_fill()
```

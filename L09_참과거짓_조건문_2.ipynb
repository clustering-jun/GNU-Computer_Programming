## L08 - 문제04의 다양한 풀이
- 밑의 풀이 외에도 다양한 소스코드가 존재할 수 있음.
 - 양수를 먼저... 음수를 먼저... 0을 먼저...

```python
n = int(input('정수 1개 입력: '))

if n > 0:
  print('양수')
else:
  if n < 0:
    print('음수')
  else:             # else의 의미: 0보다 크지도 작지도 않은 수 == 0
    print('0')
```

### 지양 해야될 코드
- if문을 3번 사용하여 변수 n보다 큰지 작은지 0인지를 여러번 비교한다.
- 필요없는 연산이 많아진다.

```python
n = int(input('정수 1개 입력: '))

if n > 0:
  print('양수')

if n < 0:
  print('음수')

if n == 0:
  print('0')
```

### if - elif - else 문법
- 구조적으로 가독성을 높이기 위해 elif라는 문법을 사용함.
- 여러 개의 조건을 순차적으로 검사하고 싶을 때 사용함.

### 문제04를 if - elif - else 문으로 풀이하면 다음과 같다.
- 실행 순서 문제
  - 변수 n이 5를 입력받았다면?
      - if n > 0:
      - print('양수')
  - 위의 두 개 코드를 실행하고 종료됨.

```python
n = int(input('정수 1개 입력: '))

if n > 0:
  print('양수')
elif n < 0:
  print('음수')
else:
  print('0')
```

## BMI 값에 따른 비만도 체크하기
- 사용자로부터 키와 체중을 입력 받아서 BMI 값을 출력하는 프로그램을 작성하시오.
- 다음을 사용하여 코딩할 것.
 - float( ) : 주어진 데이터를 실수형으로 변환하는 함수
- 입출력 설계

> 몸무게를 kg단위로 입력 하시오: 85.0

> 키를 미터 단위로 입력 하시오: 1.83

> 당신의 BMI = 25.381468541909282

image.png

```python
w = float(input('몸무게를 kg 단위로 입력하시오: '))
h = float(input('키를 미터 단위로 입력하시오: '))
BMI = w / h**2

print('당신의 BMI:', BMI)

# 판단
if BMI < 18.5:
    print('저체중')
elif BMI < 25:
    print('정상')
elif BMI < 30:
    print('과체중')
else:
    print('비만')
```

### 지양해야되는 코드
- 매우 비효율적인 코드임. 왜 그럴까?
- 크게 두 가지 이유가 있음.
 1. 독립적인 if문을 사용하여 필요 없는 비교를 하게됨.
   - 예를 들어 if-elif-else문을 사용했을 때의 경우를 생각해보면, BMI가 17.0이면 `if BMI < 18.5`와 `print('저체중')`을 실행한 후 프로그램이 종료되었겠으나 if문의 도배로 인하여 필요 없는 비교를 계속해서 하게된다.
 2. `BMI >= 25 and BMI < 30`과 같은 조건을 사용한 경우
   - 이는 elif를 통해 더 간결한 코드를 만들 수 있음. if와 elif문법을 거치며 위의 조건을 다 만족시키지 않은 값만 제어구조를 타고 내려올건데 저렇게 필요없는 논리 연산자를 활용할 필요가 없음. (가독성이나 성능면 둘 다)

```python
if BMI < 18.5:
    print('저체중')
if BMI < 25:
    print('정상')
if BMI < 30:
    print('과체중')
if BMI > 30:
    print('비만')
```

## 점수에 따라 학점 판단하기
- 다음 기준에 따라 학점을 계산하는 프로그램을 작성하시오. (단, 점수는 0점이상 100점 이하입니다.)

|점수|학점|
|------|---|
|90점 이상|A|
|80점 이상 90점 미만|B|
|70점 이상 80점 미만|C|
|60점 이상 70점 미만|D|
|60점 미만|F|

- 알고리즘: 점수입력 -> 학점판단 -> 학점출력

```python
s = float(input('점수를 입력하시오.'))

if 0 <= s <= 100:
  if s >= 90:
    print('학점 A')
  elif s >= 80:
    print('학점 B')
  elif s >= 70:
    print('학점 C')
  elif s >= 60:
    print('학점 D')
  else:
    print('학점 F')
else:
  print('점수는 0점이상 100점 이하로 입력하세요.')
```

```python
# 교수님이 제안해주신 코드

s = float(input('점수를 입력하시오.'))

if s < 0 or s > 100:
  print('점수는 0점이상 100점 이하로 입력하세요.')
elif s >= 90:
  print('학점 A')
elif s >= 80:
  print('학점 B')
elif s >= 70:
  print('학점 C')
elif s >= 60:
  print('학점 D')
else:
  print('학점 F')
```

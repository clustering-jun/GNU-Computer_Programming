## 평행한 두 선분 긋기
- 터틀 그래픽의 좌표계는 수학의 평면 좌표계와 동일하게 x, y로 생각
- 화면 중앙이 원점(0,0)
- goto(x,y) : x, y 좌표로 이동

```python
import turtle
tt = turtle.Turtle()
tt.shape("turtle")
tt.fd(100)
tt.penup()
tt.goto(0,200) # 머리 방향은 항상 고정되어 있음.
tt.pendown()
tt.fd(100)
```

## 한변이 200인 정사각형 만들기. (면색 red, 선색 blue)
- pencolor( ), fillcolor( ) 함수를 사용함.
- begin_fill( ), end_fill( ) 함수를 사용함.

```python
import turtle
t = turtle.Turtle()

t.pencolor("blue")
t.fillcolor("red")


# 채우기 시작하라
t.begin_fill() # 괄호 인자안에 "red"와 같은 색상명을 사용하면 안됨. 문법 오류

t.fd(200)
t.lt(90)
t.fd(200)
t.lt(90)
t.fd(200)
t.lt(90)
t.fd(200)
t.lt(90)

# 채우기를 끝내라.
t.end_fill()
```

```python
import turtle
t = turtle.Turtle()
t.shape("turtle")
t.color("red")

# 1사분면
t.up()
t.goto(200,200)
t.down()

t.begin_fill()

t.fd(50)
t.lt(120)
t.fd(50)
t.lt(120)
t.fd(50)
t.lt(120)

t.end_fill()


# 2사분면
t.up()
t.goto(-200,200)
t.down()

t.begin_fill()

t.fd(50)
t.lt(120)
t.fd(500)
t.lt(120)
t.fd(200)
t.lt(120)

t.end_fill()


# 3사분면
t.up()
t.goto(-200,-200)
t.down()

t.begin_fill()

t.fd(200)
t.lt(120)
t.fd(200)
t.lt(120)
t.fd(200)
t.lt(120)

t.end_fill()

# 4사분면
t.up()
t.goto(200,-200)
t.down()

t.begin_fill()

t.fd(200)
t.lt(120)
t.fd(200)
t.lt(120)
t.fd(200)
t.lt(120)

t.end_fill()
```

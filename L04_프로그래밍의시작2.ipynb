## 함수의 종류
- 메소드(Method): 특정 객체(Object)에서 호출할 수 있는 함수
 - 메소드는 클래스 내부에서 정의된 함수
 - 터틀(Turtle)은 그래픽 화면에서 거북이 모양의 커서(객체)를 조종하여 선을 그리는 방식으로 작동함. 터틀 그래픽 메소드를 사용하면 선을 긋고, 도형을 그리며, 색상을 변경하고, 거북이의 움직임을 조작할 수 있음.


- 내장 함수: print( ), str( ) 등
- 사용자 정의 함수

## 평행한 두 선분 긋기
- 터틀 그래픽의 좌표계는 수학의 평면 좌표계와 동일하게 x, y로 생각
- 화면 중앙이 원점(0,0)
- goto(x,y) : x, y 좌표로 이동
---


image.png

```python
import turtle
tt = turtle.Turtle()
tt.shape("turtle")
tt.fd(100)

tt.penup()
tt.goto(0,200) # 머리 방향은 항상 고정되어 있음.
tt.pendown()

tt.fd(100)
```

## 한변이 200인 정사각형 만들기. (면색 red, 선색 blue)
- pencolor( ), fillcolor( ) 함수를 사용함.
- begin_fill( ), end_fill( ) 함수를 사용함.
---
image.png

```python
import turtle
t = turtle.Turtle()

t.pencolor("blue")
t.fillcolor("red")


# 채우기 시작
t.begin_fill() # 인자에 "red"와 같은 색상명을 사용하면 안됨. Syntax error

t.fd(200)
t.lt(90)
t.fd(200)
t.lt(90)
t.fd(200)
t.lt(90)
t.fd(200)
t.lt(90)

# 채우기 끝
t.end_fill()
```

## 4개의 사분면에 정삼각형 그리기
- 한 변의 길이가 50인 빨간색 정삼각형 4개
---
image.png

```python
import turtle
t = turtle.Turtle()
t.shape("turtle")
t.color("red")

# 1사분면
t.up()
t.goto(200,200)
t.down()

t.begin_fill()

t.fd(50)
t.lt(120)
t.fd(50)
t.lt(120)
t.fd(50)
t.lt(120)

t.end_fill()


# 2사분면
t.up()
t.goto(-200,200)
t.down()

t.begin_fill()

t.fd(50)
t.lt(120)
t.fd(50)
t.lt(120)
t.fd(50)
t.lt(120)

t.end_fill()


# 3사분면
t.up()
t.goto(-200,-200)
t.down()

t.begin_fill()

t.fd(50)
t.lt(120)
t.fd(50)
t.lt(120)
t.fd(50)
t.lt(120)

t.end_fill()

# 4사분면
t.up()
t.goto(200,-200)
t.down()

t.begin_fill()

t.fd(50)
t.lt(120)
t.fd(50)
t.lt(120)
t.fd(50)
t.lt(120)

t.end_fill()
```

## 4개의 사분면에 다양한 색깔의 도형 그리기
- 1사분면: 빨강 삼각형 (길이 50)
- 2사분면: 노랑 정사각형 (길이 100)
- 3사분면: 파랑 삼각형 (길이 50)
- 4사분면: 보라 정사각형 (길이 100)
---
image.png

```python
import turtle
t = turtle.Turtle()
t.shape("turtle")

# 1사분면
t.color("red")
t.up()
t.goto(200,200)
t.down()

t.begin_fill()

t.fd(50)
t.lt(120)
t.fd(50)
t.lt(120)
t.fd(50)
t.lt(120)

t.end_fill()


# 2사분면
t.color("yellow")
t.up()
t.goto(-200,200)
t.down()

t.begin_fill()

t.fd(100)
t.lt(90)
t.fd(100)
t.lt(90)
t.fd(100)
t.lt(90)
t.fd(100)
t.lt(90)

t.end_fill()


# 3사분면
t.color("blue")
t.up()
t.goto(-200,-200)
t.down()

t.begin_fill()

t.fd(50)
t.lt(120)
t.fd(50)
t.lt(120)
t.fd(50)
t.lt(120)

t.end_fill()

# 4사분면
t.color("purple")
t.up()
t.goto(200,-200)
t.down()

t.begin_fill()

t.fd(100)
t.lt(90)
t.fd(100)
t.lt(90)
t.fd(100)
t.lt(90)
t.fd(100)
t.lt(90)

t.end_fill()
```

## 집그리기
### <요구사항>
- 정사각형
 - 한 변의 길이: 200
 - 선과 면 색: blue

- 정삼각형
 - 한 변의 길이: 300
 - 선과 면 색: red

### <질문사항>
- 그리기 시작할 위치는 어디가 좋을까?
 - 정사각형부터? 정삼각형부터?
   - 정답: (그리는 순서) 정사각형 -> 정삼각형
   - 이유: 만약, 정삼각형을 먼저 그리고 정사각형을 그린다면 펜 두께(5)로 인하여 삼각형이 정사각형을 침범하는 것처럼 보이기 때문임.

---
image.png

```python
import turtle
t = turtle.Turtle()
t.shape("turtle")
t.pensize(5)
t.speed(0)

# 정사각형
t.color("blue")
t.up()
t.goto(-100,-200)
t.down()

t.begin_fill()

t.fd(200)
t.lt(90)
t.fd(200)
t.lt(90)
t.fd(200)
t.lt(90)
t.fd(200)
t.lt(90)

t.end_fill()

# 정삼각형
t.color("red")
t.up()
t.goto(-150,0)
t.down()

t.begin_fill()

t.fd(300)
t.lt(120)
t.fd(300)
t.lt(120)
t.fd(300)
t.lt(120)

t.end_fill()


# 창문그리기
t.pensize(2)
t.color("white")
 # 창문 1사분면
t.up()
t.goto(-50,-50)
t.down()

t.fd(50)
t.lt(-90)
t.fd(50)
t.lt(-90)
t.fd(50)
t.lt(-90)
t.fd(50)
t.lt(-90)

 # 창문 2사분면
t.up()
t.goto(-50,-150)
t.down()

t.fd(50)
t.lt(90)
t.fd(50)
t.lt(90)
t.fd(50)
t.lt(90)
t.fd(50)
t.lt(90)

 # 창문 3 사분면
t.up()
t.goto(0,-100)
t.down()

t.fd(50)
t.lt(90)
t.fd(50)
t.lt(90)
t.fd(50)
t.lt(90)
t.fd(50)
t.lt(90)

 # 창문 4 사분면
t.up()
t.goto(0,-150)
t.down()

t.fd(50)
t.lt(90)
t.fd(50)
t.lt(90)
t.fd(50)
t.lt(90)
t.fd(50)
t.lt(90)
```

## 덧셈, 뺄셈, 곱셈, 나눗셈 연산자 활용하기
- 정수 / 정수의 값의 자료형은 실수(float)가 나옴.

```python
# 변수할당
c500 = 243
c100 = 792

# 계산
ctot = c500 + c100
m500 = c500 * 500
m100 = c100 * 100
mtot = m500 + m100
r500 = round(m500 / mtot, 2)

# 출력
print('500원 동전은 ' + str(c500) + '개\n100원 동전은 ' + str(c100) + '개\n총 개수: ' + str(ctot) + '개')

print('총 금액은 ' + str(mtot) + '원 입니다.')

print('500원 동전이 차지하는 비율은 ' + str(int(r500*100)) + '% 입니다.')
```

## 초를 몇 시간 몇 분 몇 초로 환산하기
- 중요포인트
 1. 최적화하기
    - input_data % 3600 % 60를 아래의 코드로 최적화 할 수 있음.
 2. 문자열 속에 변수의 값 삽입하여 출력하기
   - print( '%d시간 %d분 %d초' % (h, m, s))와 같은 방식으로 출력할 수 있음.

```python
input_data = int(input('초를 입력하세요: '))

h = input_data // 3600
m = (input_data % 3600) // 60
s = input_data % 60  # 최적화 부분

print(str(h)+'시간 '+str(m)+'분 '+str(s)+'초')

# 다른 출력방법!!! (암기)
print('%d시간 %d분 %d초'%(h, m, s))
```

### C 스타일 문자열 포맷팅
- 또는 **형식 지정자(format specifier)를 사용하는 문자열 포맷팅**

- %d, %s, %f는 **형식 지정자(format specifier)**라고 하며, 출력할 데이터의 **자료형(data type)**을 지정하는 역할을 함.

## 계산대 프로그램 작성하기

```python
# 입력
money = int(input("투입한 돈: "))
price = int(input("물건가격: "))

# 출력
print("거스름돈:", money - price)

print("500원 동전의 개수:", (money - price) // 500)
print("100원 동전의 개수:", (money - price) % 500 // 100)
```

## 다항식 계산하기
- -y^3 + 2x^2y의 값을 계산하여 출력하는 프로그램을 작성하시오.
- x=-1, y=3일 때, 실행 결과는 다음과 같다.
 - 21

```python
x = -1
y = 3

answer = (-y)**3 + 2*x**2*y
print('다항식의 계산 결과:', answer)
```

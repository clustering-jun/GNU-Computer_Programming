## **Function**
- 특정 기능을 수행하는 부분을 따로 떼어서 구현한 것.
- 일반적으로 반복해서 사용되는 의미있는 부분을 함수로 정의함.
- 서브루틴(subroutine), 루틴(routine), 메서드(method), 프로시저(procedure)와도 유사함.

<br>
<br>

---
> **보충 내용**

- **서브루틴 (Subroutine)**
 - 프로그램에서 반복적으로 사용되는 작업을 수행하는 독립적인 코드 블록.
 - 보통 입력을 받아서 작업을 하고, 결과를 반환하거나 어떤 동작을 수행.
 - 함수와 거의 같은 개념이지만, 반환값이 없을 수도 있음.
 - 오래된 프로그래밍 언어에서 주로 사용된 용어.

<br>

- **루틴 (Routine)**
 - 서브루틴과 거의 동일한 의미로 쓰임.
 - 일반적으로 ‘일련의 작업’을 수행하는 코드 블록을 뜻함.
 - 함수, 메서드, 프로시저 등 모두 포함하는 더 넓은 개념으로 볼 수 있음.

<br>

- **메서드 (Method)**
 - 객체 지향 프로그래밍에서 클래스 내부에 정의된 함수.
 - 특정 객체가 수행할 수 있는 동작을 나타냄.
 - 객체의 상태(속성)를 읽거나 변경하는 데 사용됨.
 - 예를 들어, car.start()에서 start가 메서드임.

<br>

- **프로시저 (Procedure)**
 - 함수와 비슷하지만, 반환값이 없는 코드 블록.
 - 작업을 수행하는 데 집중하고 결과를 직접 반환하지 않는 경우 많음.
 - 보통 절차 지향 프로그래밍에서 많이 사용됨.
 - 반환값 없는 함수라고 생각하면 됨.

- **함수를 사용되면 좋은 점**
- 함수를 사용하면 여러 면에서 코드의 효율성과 품질을 높일 수 있다.
 - 첫째, 함수는 반복되는 코드를 하나로 묶어 재사용할 수 있게 해주므로 코드의 길이를 줄이고 중복을 방지할 수 있다. 같은 작업을 여러 번 수행해야 할 때마다 함수를 호출하기만 하면 되므로 작업이 훨씬 간단하고 빠르다.
 - 둘째, 함수는 코드의 가독성을 높여준다. 코드의 목적을 명확하게 나타내는 함수 이름을 사용하면 전체 흐름을 쉽게 파악할 수 있어 다른 사람이나 미래의 자신이 코드를 이해하기 쉬워진다.
 - 셋째, 함수는 유지보수를 편리하게 해준다. 기능에 변경이 필요할 때 함수 내부 코드만 수정하면 되므로 전체 코드를 수정할 필요가 없고, 오류 발생 시에도 문제를 국한된 범위에서 해결할 수 있다.
- 이처럼 함수는 코드의 재사용성, 가독성, 유지보수성을 크게 향상시켜 프로그래밍의 핵심 도구로 사용된다.

## **인수가 없는 함수 정의하기**
- 인수가 없는 경우
 - 함수를 정의할 때 인수가 없으면, 호출할 때도 인수 없이 호출함
- 호출되면 무조건 인사하는 함수 정의

```python
def insa():
  print('안녕~')

# main program - 호출부

insa() # 함수 호출
print('프로그램 종료')
```

- **위 프로그램의 실행순서를 설명하시오.**
> 이 프로그램은 먼저 insa라는 이름의 함수를 정의하는 것으로 시작된다. 함수 정의는 실행이 아닌 준비 과정으로, insa()라는 이름으로 호출할 수 있도록 함수의 내용을 저장해 두는 역할을 한다. 함수 내부에는 print('안녕~') 명령이 포함되어 있어, 이 함수가 호출되면 "안녕~"이라는 문자열이 출력된다.

> 그 다음 줄에서는 insa() 함수를 실제로 호출하고 있다. 이 시점에서 프로그램은 위에서 정의한 insa 함수의 본문으로 이동하여 print('안녕~')를 실행하고, 그 결과로 "안녕~"이라는 문장을 화면에 출력한다.

> 함수 호출이 끝난 후 프로그램은 다음 줄인 print('프로그램 종료')를 실행한다. 이 명령은 "프로그램 종료"라는 문자열을 출력하며, 이로써 프로그램의 실행이 종료된다. 결과적으로 이 프로그램은 "안녕~"이라는 인사말을 먼저 출력한 다음, "프로그램 종료"라는 문장을 출력하며 마무리된다.

## **인수가 1개 있는 함수**

```python
def insa_call(name):
  print(name+'님 안녕~')

# main program - 호출부
name = input('이름 입력: ')
insa_call(name)

print('프로그램 종료')
```

- **참고 - Local Variable와 Global Variable**

| 구분      | 지역 변수                  | 글로벌 변수                  |
| ------- | ---------------------- | ----------------------- |
| 선언 위치   | 함수 내부                  | 함수 외부                   |
| 사용 범위   | 함수 내부에서만 사용 가능         | 프로그램 전체에서 사용 가능         |
| 생명 주기   | 함수 실행 중에만 존재           | 프로그램 실행 동안 존재           |
| 함수 내 수정 | `global` 키워드 없이는 수정 불가 | 직접 수정 가능 (단, global 필요) |

```python
def hap2(a,b):
  h = a + b
  return h

print(h)
print('프로그램 종료')
```

- h는 함수 hap2 내부에서만 선언된 지역 변수이다. 따라서 함수 외부에서는 h라는 이름의 변수를 알지 못한다. print(h)는 전역 범위에서 h를 출력하려고 하지만, 해당 범위에 h가 존재하지 않기 때문에 NameError: name 'h' is not defined 오류가 발생한다.

```python
def hap2(a,b):
  h = a + b
  return h

result = hap2(10,20)
print(result) # print(hap(10,20))
print('프로그램 종료')
```

### 합과 평균을 구하여 반환하는 함수

```python
def hap3(a,b):
  h = a + b
  avg = h / 2
  return h, avg

result = hap3(12,27) # 함수 호출 (packing)

print(type(result))
print(result)
print('프로그램 종료')
```

```python
h, a = result # unpacking
print('합:', h, '평균:', a)
```

```python
print('합:', result[0], '평균:', result[1])
```

### Exception handling을 적용한 프로그램

```python
try:
  a = int(input('수1: '))
  b = int(input('수2: '))

except:
  print('입력 오류')

else:
  result = hap3(a, b)
  print('합:', result[0], '평균:', result[1])

# 예외가 발생 여부와 관계없이 실행되는 코드
finally:
  print('프로그램 종료')
```

## **약수를 출력해 주는 함수 정의 및 호출**
- 전달받은 양의 정수 한 개에 대하여 그 수의 약수를 출력하는 함수를 정의하고 호출하여 사용해 보자.
 - 함수 이름: dividor_print(n)
 - 기능: 약수 찾기 및 출력 기능
 - 반환값 없음.

```python
n = int(input('입력:'))

def dividor_print(n):
  for i in range(1,n+1):
    if n % i == 0:
      print(i, end = ' ')

dividor_print(n)
```

<br>

 - 함수 이름: dividor_print(n)
 - 기능: 약수 찾기 및 출력 기능
 - 반환값 없음.

```python
def dividor_cnt(n):
    a = []
    for i in range(1, n + 1):
        if n % i == 0:
            a.append(i)
    return len(a)
```

- 약수 개수를 이용하여 소수 여부 판단하기

```python
n = int(input('입력:'))

def dividor_print(n):
  a = []
  for i in range(1,n+1):
    if n % i == 0:
      a.append(i)
  return len(a)

if dividor_print(n) == 2:
  print('소수')
else:
  print('소수아님')
```

## **전체 소스코드**

```python
# 약수를 출력해 주는 함수 정의
def dividor_print(n):
    print('약수:', end=' ')
    for i in range(1, n + 1):
        if n % i == 0:
            print(i, end=' ')
    print()

# 약수의 개수를 세어 주는 함수 정의
def dividor_cnt(n):
    a = []
    for i in range(1, n + 1):
        if n % i == 0:
            a.append(i)
    return len(a)

# Main program (함수 호출)
n = int(input('입력:'))

# 약수 출력
dividor_print(n)

# 약수 개수 출력
print('약수의 개수:', dividor_cnt(n))

# 소수 여부 출력
if dividor_cnt(n) == 2:
    print('소수입니다.')
else:
    print('소수가 아닙니다.')
```

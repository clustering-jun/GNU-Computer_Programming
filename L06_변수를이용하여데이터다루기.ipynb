# L06. 변수를이용하여데이터다루기
## 오륜기 그리기
- teleport( ), circle( ) 함수 사용
---
image.png

```python
import turtle
t = turtle.Turtle()
t.shape("turtle")
t.pensize(10)
t.speed(0)

t.color('black')
t.circle(100)

t.teleport(-220,0)
t.color('blue')
t.circle(100)

t.teleport(220,0)
t.color('red')
t.circle(100)

t.teleport(-110,-100)
t.color('yellow')
t.circle(100)

t.teleport(110,-100)
t.color('green')
t.circle(100)

```

## 데이터의 평균을 구하는 프로그램
- 자연수가 들어올 동안 데이터의 평균을 구하는 프로그램을 작성하시오.
- 0 이하의 값이 들어오면 더 이상 입력 받지 않는다. 데이터가 몇 개 들어올 지 알 수 없으며, 언제 0이하의 값이 들어올 지 알 수 없을 때, while문을 이용하여 구현하시오.

 - `Ctrl + C`: 현재 파이썬 프로그램을 강제 중단
   - 내부 처리	KeyboardInterrupt 예외 발생


```python
h = 0
c = 0

while True:
  n = int(input('자연수 입력 (0이하 입력되면 종료): '))

  if n > 0:
    h += n
    c += 1
  else:
    break

if c > 0:
  ave = h / c
  print(h, c, ave)
else:
  print('데이터가 없습니다.')
```

- 첫 입력이 0일 경우, c = 0이므로 h / c는 ZeroDivisionError 발생 가능 → 이를 if c > 0: 조건으로 방지.
- 실수 입력 (예: 0.2, 0.02) 시 int()에서 ValueError 발생 → 입력값에 대한 예외처리 필요.
- 입력값 검증이나 예외 상황이 많으므로, 단순 조건문보다는 try-except를 활용한 예외 처리가 효율적.

## (참고) f-string
- formatted string literal
- 문자열을 보다 간결하고 가독성 좋게 포맷팅하는 방법 (python 3.6 도입)
- 문자열 앞에 f 또는 F를 붙이면, 중괄호 {} 안에 변수나 표현식을 직접 쓸 수 있음.

```python
for i in range(1,10):
  print(f'2 * {i} = {2*i}')
```

## [Data Structure] 데이터 나열하여 저장하기: 리스트(List)
- 여러 개의 데이터를 하나의 그룹으로 동일한 변수에 순차적으로 저장
- 여러 내장 함수를 통해 다양한 기능을 수행

<br>

- 리스트의 색인(index): 리스트 타입 데이터에는 자동으로 index가 생김.
 - 기본 색인은 0부터 작하여 정수 형태로 1씩 증가
 - 끝에서부터 접근을 위해 음수 색인도 제공
   - 리스트의 색인 범위 오류: IndexError: string index out of range

<br>

- 리스트의 길이 확인: `len( )`함수로 요소의 개수 확인

```python
a = [3, 1, 5, 2]

print(a) # [3, 1, 5, 2]
print(len(a)) # 4

print(a[0]) # 3
print(a[-1]) # 2
print(a[5]) # IndexError: string index out of range


print(type(a)) # <class 'list'>
```

```python
import turtle

colors = ['red', 'purple', 'blue', 'green', 'yellow', 'orange']
t = turtle.Turtle()

turtle.bgcolor('black')
t.speed(0)
t.width(3)
length = 10

while length < 500:
  t.forward(length)
  t.pencolor(colors[ length % 6 ] )
  t.right(89)
  length += 5
```

- `length`가 500보다 작을 동안 계속해서 루프를 반복
- 현재 방향으로 length만큼 거북이를 이동시키며 선을 그림.
- 처음에는 10픽셀 직선이 그려짐.

<br>
- `t.pencolor(colors[ length % 6 ])`
 - 펜 색상을 설정. 색상은 length 값을 6으로 나눈 나머지를 인덱스로 하여 colors 리스트에서 선택됨.

- black 이후 `yellow → green → blue → purple → red → orange`를 반복함.

## 난수와 리스트의 활용
- 한 변의길이가 50인 정사각형 20개
- 임의의 위치에 임의의 색의 정사각형을 그리는 프로그램을 구현하시오.

```python
import turtle
import random

r = random.Random()
t = turtle.Turtle()
t.speed(0)

color = ['red', 'blue', 'yellow', 'green', 'purple', 'orange', 'pink', 'black']

for i in range(20):
    x = r.randint(-800,800)
    y = r.randint(-400,400)
    c = r.randint(0,7)

    t.color(color[c])
    t.penup()
    t.goto(x,y)
    t.pendown()

    t.begin_fill()

    t.fd(50)
    t.lt(90)
    t.fd(50)
    t.lt(90)
    t.fd(50)
    t.lt(90)
    t.fd(50)
    t.lt(90)

    t.end_fill()
```
